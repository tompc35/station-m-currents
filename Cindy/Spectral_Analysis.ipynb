```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
import scipy.signal as signal
```

```python
data = dates_2016_2018['Easting (cm/sec)']
```

```python
record_length = (date_time_number[-1]-date_time_number[0])
#top_time = (2*np.pi)/(1/24)
print(record_length)

top = 24*(np.pi*2)
freq = np.linspace(((1/record_length)*(np.pi*2)), top, 100)
```

```python
top2 = np.log10(top)
first = np.log10((1/record_length)*(np.pi*2))

freq3 = np.logspace(first, top2, 1000)
```

```python
pgram = signal.lombscargle(date_time_number, data, freq3)

freq4 = freq3/(2*np.pi)
```

```python
plt.figure()
#plt.plot(freq, pgram)
plt.loglog(freq4,pgram)
plt.savefig('2016_spectral_analysis.png')
```

```python
#creating periodogram without using the lomb-scargle technique
hourly_2014_2015 = dates_2014_2015.resample('1h').mean()
data_2014 = dates_2014_2015['Easting (cm/sec)']
N_2014 = len(data_2014)
deltat = 1/24
f_2014 = np.fft.fftfreq(N_2014,deltat)
fft_2014 = np.fft.fft(data_2014)
S_2014 = (2/N_2014)*deltat*np.abs(fft_2014)**2 

#fft periodogram
plt.figure(figsize = (10,10))
plt.subplot(2,1,1)
plt.loglog(f_2014[0:round(N_2014/2)],S_2014[0:round(N_2014/2)])
plt.title('2014-2015 Spectral Analysis')
#Lomb-Scargle periodogram
plt.subplot(2,1,2)
plt.loglog(freq_2014_2, pgram_2014)
#plt.title('2014-2015 Lonb-Scargle Spectral Analysis')

```
